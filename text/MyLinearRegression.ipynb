```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

X_Train = pd.read_csv("./data/X_train.csv")
y_Train = pd.read_csv("./data/y_train.csv",header=None)
X_Test = pd.read_csv("./data/X_test.csv")

display(X_Train)
print(y_Train)
```

```python
X_train=X_Train[["平均湿度"]].values
X_test=X_Test[["平均湿度"]].values

T_train=X_Train[["降水量の合計"]].values
T_test=X_Test[["降水量の合計"]].values

y_train=np.array(y_train)
print(X_train.shape)
print(y_train.shape)
```

```python
X_train=X_train-X_train.mean()
X_test=X_test-X_test.mean()

T_train=T_train-T_train.mean()
T_test=T_test-T_test.mean()

y_train=y_train-y_train.mean()
```

```python
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.utils.validation import check_X_y, check_is_fitted, check_array

class MyLinearRegression(BaseEstimator, RegressorMixin):
    def fit(self, X, y):
        X = check_array(X)
        self.coef_ = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, y)) 
        
        return self
    
    def predict(self,X):
        check_is_fitted(self, 'coef_')
        X = check_array(X)
        y = np.dot(X, self.coef_)
        
        return y
```

```python
clf = MyLinearRegression()
clf.fit(X_train, T_train)

# 決定係数
print("決定係数: ", clf.score(X_train, T_train))


fig, axes = plt.subplots(ncols=2, figsize = (12,4))

axes[0].scatter(X_train,T_train,marker=".")
axes[0].plot(X_train, clf.predict(X_train), color="red")
axes[0].set_title("train")

axes[1].scatter(X_test, T_test, marker=".")
axes[1].plot(X_test, clf.predict(X_test), color="red")
axes[1].set_title("test")
fig.show()
```
